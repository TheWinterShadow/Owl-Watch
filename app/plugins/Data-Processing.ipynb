```python
us_cases_df = pd.read_csv('https://web.netsecure.dev/files/covid/US_cases.csv')
us_deaths_df = pd.read_csv('https://web.netsecure.dev/files/covid/US_deaths.csv')
```

```python
us_cases_df.drop(['UID', 'iso2', 'iso3', 'code3', 'FIPS', 'Country_Region', 'Combined_Key'], inplace=True, axis=1)
us_cases_df.rename(columns={'Admin2':'County', 'Province_State':'State', 'Long_':'Long'}, inplace=True)
```

```python
us_cases_df['Index'] = us_cases_df['State'] + '-' + us_cases_df['County']
us_cases_df.set_index('Index', inplace=True)
us_cases_df.index.name = None
```

```python
dates = list(us_cases_df.columns.drop(['County', 'State', 'Lat', 'Long']))
us_cases_df['Total'] = us_cases_df[dates].sum(axis=1)
```

```python
us_cases_df.sort_values(by='Total', ascending=False, inplace=True)
```
